### Good Quality Images

```python
im = array(Image.open(r'E:\Downloads\iqa_assignment\good_quality\47.png').convert('L'))
im = im / 255.0
# calculating MSCN coefficients
mu = cv2.GaussianBlur(im, (7, 7), 1.166)
mu_sq = mu * mu
sigma = cv2.GaussianBlur(im*im, (7, 7), 1.166)
sigma = (sigma - mu_sq)**0.5
# structdis is the MSCN image
structdis = im - mu
structdis /= (sigma + 1.0/255)

hist(structdis.flatten(), 128)
plt.show()
```
